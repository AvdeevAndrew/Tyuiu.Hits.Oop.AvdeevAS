```
using System;

public class Person
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string Gender { get; set; }
    public string Address { get; set; }

    public Person(string name, int age, string gender, string address)
    {
        Name = name;
        Age = age;
        Gender = gender;
        Address = address;
    }

    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Пол: {Gender}, Адрес: {Address}";
    }

    public virtual void SayHello()
    {
        Console.WriteLine($"Здравствуйте, меня зовут {Name}.");
    }

    public void Greet(Person other)
    {
        Console.WriteLine($"{Name} приветствует {other.Name}.");
    }

    public void Celebrate()
    {
        Console.WriteLine($"{Name} празднует день рождения!");
    }
}

public interface IWorkable
{
    void Work();
    decimal GetIncome();
}

public interface IStudiable
{
    void Study();
    string GetEducationLevel();
}

public interface ITeachable
{
    void Teach();
    void GradeStudent(Student student, int grade);
}

public class Student : Person, IStudiable
{
    public string University { get; set; }
    public int Course { get; set; }
    public double GPA { get; set; }
    public string Major { get; set; }

    public Student(string name, int age, string gender, string address, string university, int course, double gpa, string major)
        : base(name, age, gender, address)
    {
        University = university;
        Course = course;
        GPA = gpa;
        Major = major;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}, Университет: {University}, Курс: {Course}, Средний балл: {GPA}, Специальность: {Major}";
    }

    public override void SayHello()
    {
        Console.WriteLine($"Здравствуйте, меня зовут {Name}. Я учусь на {Course} курсе по специальности {Major}.");
    }

    public void Study()
    {
        Console.WriteLine($"{Name} изучает материалы по специальности {Major}.");
    }

    public string GetEducationLevel()
    {
        return $"Высшее образование, {Course} курс";
    }

    public void TakeExam(string subject)
    {
        Console.WriteLine($"{Name} сдает экзамен по предмету {subject}.");
    }
}

public class GraduateStudent : Student
{
    public string ThesisTopic { get; set; }
    public string Supervisor { get; set; }

    public GraduateStudent(string name, int age, string gender, string address, string university, int course, double gpa, string major, string thesisTopic, string supervisor)
        : base(name, age, gender, address, university, course, gpa, major)
    {
        ThesisTopic = thesisTopic;
        Supervisor = supervisor;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}, Тема диссертации: {ThesisTopic}, Научный руководитель: {Supervisor}";
    }

    public void DefendThesis()
    {
        Console.WriteLine($"{Name} защищает диссертацию на тему: {ThesisTopic}.");
    }
}

public class Employee : Person, IWorkable
{
    public string Company { get; set; }
    public decimal Salary { get; set; }
    public string Position { get; set; }
    public int WorkExperience { get; set; }

    public Employee(string name, int age, string gender, string address, string company, decimal salary, string position, int workExperience)
        : base(name, age, gender, address)
    {
        Company = company;
        Salary = salary;
        Position = position;
        WorkExperience = workExperience;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}, Компания: {Company}, Зарплата: {Salary:C}, Должность: {Position}, Опыт работы: {WorkExperience} лет";
    }

    public void Work()
    {
        Console.WriteLine($"{Name} работает в должности {Position} в компании {Company}.");
    }

    public decimal GetIncome()
    {
        return Salary;
    }

    public void GetPromotion(string newPosition, decimal newSalary)
    {
        Console.WriteLine($"{Name} получил повышение! Новая должность: {newPosition}, Новая зарплата: {newSalary:C}");
        Position = newPosition;
        Salary = newSalary;
    }
}

public class Teacher : Person, ITeachable, IWorkable
{
    public string Subject { get; set; }
    public int Experience { get; set; }
    public string School { get; set; }
    public decimal Salary { get; set; }

    public Teacher(string name, int age, string gender, string address, string subject, int experience, string school, decimal salary)
        : base(name, age, gender, address)
    {
        Subject = subject;
        Experience = experience;
        School = school;
        Salary = salary;
    }

    public override string GetInfo()
    {
        return $"{base.GetInfo()}, Предмет: {Subject}, Опыт: {Experience} лет, Школа: {School}, Зарплата: {Salary:C}";
    }

    public override void SayHello()
    {
        Console.WriteLine($"Здравствуйте, меня зовут {Name}. Я преподаю {Subject} в {School}.");
    }

    public void Teach()
    {
        Console.WriteLine($"{Name} ведет урок по предмету {Subject}.");
    }

    public void GradeStudent(Student student, int grade)
    {
        Console.WriteLine($"{Name} поставил оценку {grade} студенту {student.Name}.");
    }

    public void Work()
    {
        Console.WriteLine($"{Name} работает учителем по предмету {Subject}.");
    }

    public decimal GetIncome()
    {
        return Salary;
    }
}

Person person = new Person("Андрей", 20, "Мужской", "ул. Ленина, 10");
person.SayHello();
Console.WriteLine(person.GetInfo());
person.Celebrate();
Console.WriteLine();

Student student = new Student("Мария", 20, "Женский", "ул. Пушкина, 5", "ТГУ", 2, 4.5, "Информатика");
student.SayHello();
Console.WriteLine(student.GetInfo());
student.Study();
Console.WriteLine($"Уровень образования: {student.GetEducationLevel()}");
student.TakeExam("Базы данных");
Console.WriteLine();

GraduateStudent gradStudent = new GraduateStudent("Алексей", 24, "Мужской", "ул. Гагарина, 15", "МГУ", 6, 4.8, "Искусственный интеллект", "Машинное обучение в медицине", "Профессор Иванов");
gradStudent.SayHello();
Console.WriteLine(gradStudent.GetInfo());
gradStudent.Study();
gradStudent.DefendThesis();
Console.WriteLine();

Employee employee = new Employee("Иван", 35, "Мужской", "ул. Советская, 20", "ТехноКорп", 75000, "Менеджер проектов", 10);
employee.SayHello();
Console.WriteLine(employee.GetInfo());
employee.Work();
Console.WriteLine($"Доход: {employee.GetIncome():C}");
employee.GetPromotion("Старший менеджер", 95000);
Console.WriteLine();

Teacher teacher = new Teacher("Ольга", 40, "Женский", "ул. Чехова, 8", "Математика", 15, "Гимназия №1", 55000);
teacher.SayHello();
Console.WriteLine(teacher.GetInfo());
teacher.Teach();
teacher.Work();
Console.WriteLine($"Доход: {teacher.GetIncome():C}");
teacher.GradeStudent(student, 5);
Console.WriteLine();

person.Greet(student);
student.Greet(employee);
employee.Greet(teacher);
teacher.Greet(gradStudent);
```
